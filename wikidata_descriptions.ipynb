```python
from wikidata.client import Client
```

```python
client = Client()  # doctest: +SKIP
```

```python
# entity
entity = client.get('Q20145', load=True)
print(entity.description)
```

```python
# Images

from IPython.core.display import Image, display
image_prop = client.get('P18')

image = entity[image_prop]
print(image.image_resolution)

target_height = 200
scale_factor = round(image.image_resolution[1] / target_height,0)

width = (image.image_resolution[0]) / scale_factor
height = (image.image_resolution[1]) / scale_factor

print(image)
display(Image(image.image_url, width=width, height=height))
```
